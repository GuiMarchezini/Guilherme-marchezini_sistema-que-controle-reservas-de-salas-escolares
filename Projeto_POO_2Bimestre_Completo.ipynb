## PARTE 1 ‚Äì Classe Principal e Objetos

**Tarefas:**
- Definir e criar a classe principal com pelo menos 3 atributos.
- Implementar o m√©todo construtor (`__init__`).
- Criar pelo menos 2 objetos da classe principal.
- Implementar um m√©todo de exibi√ß√£o de dados.

**C√≥digo:**

```python
class ReservaSala_GuilhermeMarchezini:
    def __init__(self, sala, responsavel, horario):
        self.sala = sala
        self.responsavel = responsavel
        self.horario = horario

    def exibir_reserva(self):
        print("=== Dados da Reserva ===")
        print(f"Sala: {self.sala}")
        print(f"Respons√°vel: {self.responsavel}")
        print(f"Hor√°rio: {self.horario}")
        print("========================\n")


# Lista para armazenar as reservas
reservas = []


def adicionar_reserva():
    sala = input("Informe o nome da sala: ")
    responsavel = input("Informe o nome do respons√°vel: ")
    horario = input("Informe o hor√°rio da reserva: ")
    nova_reserva = ReservaSala_GuilhermeMarchezini(sala, responsavel, horario)
    reservas.append(nova_reserva)
    print("Reserva adicionada com sucesso!\n")


def remover_reserva():
    if not reservas:
        print("N√£o h√° reservas para remover.\n")
        return

    print("Reservas existentes:")
    for i, reserva in enumerate(reservas):
        print(f"{i + 1}. Sala: {reserva.sala}, Respons√°vel: {reserva.responsavel}, Hor√°rio: {reserva.horario}")

    try:
        indice = int(input("Digite o n√∫mero da reserva que deseja remover: ")) - 1
        if 0 <= indice < len(reservas):
            removida = reservas.pop(indice)
            print(f"Reserva da sala {removida.sala} removida com sucesso!\n")
        else:
            print("N√∫mero inv√°lido.\n")
    except ValueError:
        print("Entrada inv√°lida. Digite apenas n√∫meros.\n")


def visualizar_reservas():
    if not reservas:
        print("Nenhuma reserva cadastrada.\n")
    else:
        for reserva in reservas:
            reserva.exibir_reserva()


def menu():
    while True:
        print("===== Sistema de Reservas =====")
        print("1. Adicionar reserva")
        print("2. Remover reserva")
        print("3. Visualizar reservas")
        print("4. Sair")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            adicionar_reserva()
        elif opcao == "2":
            remover_reserva()
        elif opcao == "3":
            visualizar_reservas()
        elif opcao == "4":
            print("Encerrando o sistema. At√© mais!")
            break
        else:
            print("Op√ß√£o inv√°lida. Tente novamente.\n")


# Iniciar o menu interativo
if __name__ == "__main__":
    menu(1)
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**

- Criar m√©todos `get` e `set` para o atributo privado.
- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python

class ReservaSala_GuilhermeMarchezini:
    def __init__(self, sala, responsavel, horario):
        self.__sala = sala  # Atributo privado
        self.responsavel = responsavel
        self.horario = horario

   
    def get_sala(self):
        return self.__sala

    
    def set_sala(self, nova_sala):
        if isinstance(nova_sala, str) and nova_sala.strip() != "":
            self.__sala = nova_sala
        else:
            print("Erro: Nome da sala inv√°lido.")

    
    def verificar_disponibilidade(self, horario_desejado):
        if self.horario == horario_desejado:
            return "Hor√°rio indispon√≠vel."
        else:
            return "Hor√°rio dispon√≠vel."

    
    def exibir_reserva(self):
        print("=== Dados da Reserva ===")
        print(f"Sala: {self.__sala}")
        print(f"Respons√°vel: {self.responsavel}")
        print(f"Hor√°rio: {self.horario}")
        print("========================\n")
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3


class ReservaSala_GuilhermeMarchezini:
    def __init__(self, sala, responsavel, horario):
        self.sala = sala
        self.responsavel = responsavel
        self.horario = horario

    def exibir_reserva(self):
        print("=== Reserva de Sala ===")
        print(f"Sala: {self.sala}")
        print(f"Respons√°vel: {self.responsavel}")
        print(f"Hor√°rio: {self.horario}")
        print("=======================\n")


class ReservaLaboratorio_GuilhermeMarchezini(ReservaSala_GuilhermeMarchezini):
    def __init__(self, sala, responsavel, horario, equipamentos):
        super().__init__(sala, responsavel, horario)
        self.equipamentos = equipamentos

    
    def exibir_reserva(self):
        print("=== Reserva de Laborat√≥rio ===")
        print(f"Sala: {self.sala}")
        print(f"Respons√°vel: {self.responsavel}")
        print(f"Hor√°rio: {self.horario}")
        print("Equipamentos:")
        for item in self.equipamentos:
            print(f" - {item}")
        print("==============================\n")


conn = sqlite3.connect('reservas.db')
cursor = conn.cursor()


cursor.execute('''
    CREATE TABLE IF NOT EXISTS reservas (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        sala TEXT,
        responsavel TEXT,
        horario TEXT
    )
''')
conn.commit()


def inserir_reserva(sala, responsavel, horario):
    cursor.execute("INSERT INTO reservas (sala, responsavel, horario) VALUES (?, ?, ?)", (sala, responsavel, horario))
    conn.commit()
    print("Reserva inserida com sucesso!\n")


def listar_reservas():
    cursor.execute("SELECT * FROM reservas")
    resultados = cursor.fetchall()
    if resultados:
        for reserva in resultados:
            print(f"ID: {reserva[0]} | Sala: {reserva[1]} | Respons√°vel: {reserva[2]} | Hor√°rio: {reserva[3]}")
    else:
        print("Nenhuma reserva encontrada.")
    print()


def excluir_reserva(id_reserva):
    cursor.execute("DELETE FROM reservas WHERE id = ?", (id_reserva,))
    conn.commit()
    print(f"Reserva com ID {id_reserva} exclu√≠da.\n")




reserva1 = ReservaSala_GuilhermeMarchezini("Sala 101", "Prof. Ana", "14:00 - 15:00")
reserva2 = ReservaLaboratorio_GuilhermeMarchezini("Lab 305", "Prof. Jo√£o", "10:00 - 12:00", ["Projetor", "Computadores"])


reserva1.exibir_reserva()
reserva2.exibir_reserva()


inserir_reserva(reserva1.sala, reserva1.responsavel, reserva1.horario)
inserir_reserva(reserva2.sala, reserva2.responsavel, reserva2.horario)


listar_reservas()


excluir_reserva(1)


listar_reservas()


conn.close()
```
